## Câu 4

- Vẽ hist cho các cột dữ liệu số
- Phân tích các khía cạnh thú vị của hist (2-3 câu)

```python
def draw_hist(colName, bins=20, figsize=(8,5), rotation=0):
    plt.figure(figsize=figsize)
    plt.xticks(rotation=rotation)
    plt.grid(b=True, zorder=0)
    plt.hist(df[colName], bins=bins, density=True, zorder=5)
    plt.xlabel(colName)
    plt.ylabel('prob')
    plt.title(f'Histogram of {colName}, bins = {bins}')
```

```python
# mpg
draw_hist('mpg')
```

```python
# cylinders
draw_hist('cylinders', bins=5)
```

```python
# displacement
draw_hist('displacement', bins=18)
```

```python
# horsepower
draw_hist('horsepower')
```

```python
# weight
draw_hist('weight')
```

```python
# acceleration
draw_hist('acceleration', bins=19)
```

```python
# model	
draw_hist('model', bins=13)
```

```python
# origin
draw_hist('origin', bins=3)
```

## Câu 5

- Vẽ scatter weight và MPG
- Tính hệ số tương quan

```python
plt.figure(figsize=(10,7))
plt.scatter(df['weight'], df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.title('Scatter plot giữa Weight và MPG')
```

```python
cov = np.cov(df['weight'], df['mpg'], ddof=0)
cov[0][1]
```

- Nhận xét: 
    - Hệ số tương cao cao vcl và âm chứng tỏ giữa 2 cái này tồn tại quan hệ nghịch biến
    - Điều này là dễ hiểu vì khi tăng khối lượng xe và giữ nguyên lượng xăng (1 gallon) thì mpg bao sure là giảm

## Câu 6: (BB)
  - Vẽ scatter plot year vs cylinder. Cộng random noise vào để đẹp hơn

```python
plt.figure(figsize=(10,7))
year = df['model'] + 1900 + np.random.rand(len(df['model'])) * 2
cylinders = df['cylinders'] + np.random.rand(len(df['cylinders']))

plt.scatter(year, cylinders)
plt.xlabel('year')
plt.ylabel('cylinders')
plt.title('Scatter plot giữa Year và cylinders')
```
